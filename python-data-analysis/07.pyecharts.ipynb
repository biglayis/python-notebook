# pyecharts介绍<a id="1"></a>

Echarts 是百度开发的一个开源前端图表库，2018年1月16日，ECharts 进入 Apache Incubator 进行孵化，目前已经是 Apache 软件基金会的顶级项目。凭借着良好的交互性和精巧的图表设计，ECharts 得到了众多开发者的认可，而 pyecharts 就是基于 Python 语言对 ECharts 进行了包装，让 Python 开发者也可以使用 ECharts 绘制外观精美且交互性强的统计图表。

## 特性<a id="1.1"></a>

* 简洁的 API 设计，使用如丝滑般流畅，支持链式调用
* 囊括了 30+ 种常见图表，应有尽有
* 支持主流 Notebook 环境，Jupyter Notebook 和 JupyterLab
* 可轻松集成至 Flask，Django 等主流 Web 框架
* 高度灵活的配置项，可轻松搭配出精美的图表
* 详细的文档和示例，帮助开发者更快的上手项目
* 多达 400+ 地图文件以及原生的百度地图，为地理数据可视化提供强有力的支持

## 官方文档<a id="1.2"></a>

* 中文文档：https://pyecharts.org/#/zh-cn/
* 英文文档：https://pyecharts.org/#/en-us/
* 示例 Example：https://gallery.pyecharts.org/#/

# 配置项<a id="3"></a>

图形的参数配置是数据可视化的基础，pyecharts中的图形配置项分为全局配置项和系列配置项，目前官方文档中共包含了22个全局配置项和17个系列配置项。

全局配置项可以通过`set_global_options`设置，主要包括：
* 基本元素配置项：主要包括`InitOpts`（初始化：含画布高/宽度、网页标题、图表主题、背景颜色等）、`ToolBoxFeatureOpts`（工具箱工具：含保存图片、配置项还原等）、`ToolboxOpts`（工具箱：含是否显示工具栏组件、工具栏icon的布局朝向等）、`TitleOpts`（标题：含主/副标题文本等）、`LegendOpts`（图例：含图例类型等）等；
* 坐标轴配置项：主要包括`AxisLineOpts`（坐标轴轴线：含是否显示坐标轴、轴线是否在另一个轴的0刻度上等）、`AxisTickOpts`（坐标轴刻度：含是否显示刻度、坐标轴刻度长度等）、`AxisOpts`（坐标轴：含坐标轴类型、名称、刻度最小/大值等）等；
* 原生图形配置项：主要包括`GraphicGroup`、`GraphicItem`等；

系列配置项可以通过`set_series_options`设置，主要包括：  
* 样式类配置项：主要包括`ItemStyleOpts`（图元样式：含图形颜色、阴线图形颜色等）、`TextStyleOpts`（文字样式：含文字颜色、字体、大小等）、`LabelOpts`（标签：含是否显示标签、标签文字的设置等）等；
* 标记类配置项：主要包括`MarkPointItem`（标记点数据项）、`MarkLineItem`（标记线数据项）等；
* 其他类配置项：主要包括`EffectOpts`（涟漪特效）、`AreaStyleOpts`（区域填充样式）等；

设置配置项需要导入pyecharts的pyecharts模块：

```python
import pyecharts.options as opts
```

## 柱状图<a id="4.1"></a>

如果要绘制柱图，可以使用charts模块的Bar对象

```python
from pyecharts.charts import Bar
```

Bar有一个初始化参数`init_opts`，用于配置画布高/宽度、网页标题、图表主题、背景颜色等，传入的是全局配置项中的`InitOpts`对象

Bar对象常用的方法如下：

|方法|描述|参数|
|:---:|:---:|:---:|
|add_xaxis()|添加x轴信息|xaxis_data:参数为x轴要展示的数据|
|add_yaxis()|添加y轴信息|series_name：系列名称<br>y_axis：y轴要展示的数据<br>stack：数据堆叠<br>color：数据标签的颜色|
|set_series_opts()|系列配置项|label_opts：标签配置<br>axisline_opts：坐标轴轴线配置<br>markpoint_opts：标记点配置<br>markline_opts：标记线配置<br>itemstyle_opts：图元样式配置|
|set_global_opts()|全局配置项|title_opts：标题配置<br>legend_opts：图例配置<br>tooltip_opts：提示框配置<br>xaxis_opts：x轴配置<br>yaxis_opts：y轴配置|
|render()|生成 HTML|path:设置保存位置|
|render_notebook()|在notebook中渲染图表||

下面是一个简单的柱图示例：

```python
from pyecharts.faker import Faker    # 导入pyecharts自带数据集Faker

# 准备数据
x = Faker.clothes    # ['衬衫', '毛衣', '领带', '裤子', '风衣', '高跟鞋', '袜子']
y1 = Faker.values()    # Faker的values方法可以随机生成20-150之间的7个数字的列表
y2 = Faker.values()

# 新建一个Bar对象
bar = Bar(init_opts=opts.InitOpts(width='600px', height='400px'))    # 创建Bar对象并初始化，宽度和高度分别为600，400
bar.add_xaxis(x)                                                     # 添加x轴数据
bar.add_yaxis(series_name='一季度',y_axis=y1)                        # 添加y轴数据，并设置y轴的系列名称，用于图例和文字提示
bar.add_yaxis(series_name='二季度',y_axis=y2) 
bar.set_global_opts(xaxis_opts=opts.AxisOpts(name='服装'))           # 设置x轴名称

# 在jupyter notebook中渲染图形
bar.render_notebook()    
```

因为pyecharts支持链式调用，所以上述代码也可以写成：

```python
bar = (
    Bar(init_opts=opts.InitOpts(width='600px', height='400px'))    
    .add_xaxis(x)                               
    .add_yaxis(series_name='一季度',y_axis=y1)    
    .add_yaxis(series_name='二季度',y_axis=y2) 
    .set_global_opts(xaxis_opts=opts.AxisOpts(name='服装'))    
)

bar.render_notebook()   
```

通过`stack`参数可以设置堆积柱图：

```python
bar = (
    Bar(init_opts=opts.InitOpts(width='600px', height='400px'))    
    .add_xaxis(x)                               
    .add_yaxis(series_name='一季度',y_axis=y1,stack='y1')    # 不同系列设置相同的stack名称，可以使其堆积
    .add_yaxis(series_name='二季度',y_axis=y2,stack='y1') 
    .set_global_opts(xaxis_opts=opts.AxisOpts(name='服装'))    
)

bar.render_notebook()   
```

`reversal_axis()`方法可以使x轴和y轴翻转，这样我们就得到了水平条形图：

```python
bar = (
    Bar(init_opts=opts.InitOpts(width='600px', height='400px'))    
    .add_xaxis(x)                               
    .add_yaxis(series_name='一季度',y_axis=y1,stack='y1')    # 不同系列设置相同的stack名称，可以使其堆积
    .add_yaxis(series_name='二季度',y_axis=y2,stack='y1') 
    .set_global_opts(xaxis_opts=opts.AxisOpts(name='服装'))    
    .set_series_opts(label_opts=opts.LabelOpts(position='right'))    # 将数据标签设置在右侧
    .reversal_axis()    # x，y轴翻转
)

bar.render_notebook()   
```

通过`stack`参数，还可以绘制瀑布图：

```python
x_data = [f"11月{str(i)}日" for i in range(1, 12)]      # 日期
y_in = [900, 345, 393, "-", "-", 135, 178, 286, "-", "-", "-"]    # 收入
y_out = ["-", "-", "-", 108, 154, "-", "-", "-", 119, 361, 203]   # 支出
y_total = [0, 900, 1245, 1530, 1376, 1376, 1511, 1689, 1856, 1495, 1292]    # 累计总收入


bar = (
    Bar()
    .add_xaxis(x_data)    # 添加x轴数据
    # 实际上是通过将柱子颜色设置为透明的方法来实现瀑布图，必须先添加总收入数据，先添加的数据堆积在下方
    .add_yaxis(     
        series_name="",
        y_axis=y_total,
        stack="总量",
        itemstyle_opts=opts.ItemStyleOpts(color="rgba(0,0,0,0)")   
    )
    .add_yaxis(series_name="收入", y_axis=y_in, stack="总量")      # 添加收入数据
    .add_yaxis(series_name="支出", y_axis=y_out, stack="总量")     # 添加支出数据
)

bar.render_notebook()
```

## 折线图<a id="4.2"></a>

如果要绘制柱图，可以使用charts模块的`Line`对象

```python
from pyecharts.charts import Line
```

Line对象同样有初始化参数`init_opts`，以及`add_xaxis()`，`add_yaxis()`等方法，方法的属性有部分不同，具体可以参考官方文档。

下面是一个折线图的示例：

```python
x = Faker.week    # ['周一', '周二', '周三', '周四', '周五', '周六', '周日']
y1 = [11, 11, 15, 13, 12, 13, 10]
y2 = [1, -2, 2, 5, 3, 2, 0]

line = (
    Line()
    .add_xaxis(x)
    .add_yaxis(series_name='最高气温' ,y_axis=y1)
    .add_yaxis(series_name='最低气温',y_axis=y2)
    .set_global_opts(
        title_opts=opts.TitleOpts(title="未来一周气温变化"),    # 添加图表标题
        xaxis_opts=opts.AxisOpts(type_="category", boundary_gap=False)   # 将x轴设置为分类轴，数据标签不间隔显示
    )
)

line.render_notebook()
```

我们可以通过系列配置项中的`MarkPointOpts`标出图中的最高点、最低点、平均点，通过`MarkLineItem`添加最高线、最低线、平均线：

```python
line = (
    Line()
    .add_xaxis(x)
    .add_yaxis(
        series_name='最高气温' ,
        y_axis=y1,
        markpoint_opts=opts.MarkPointOpts(                         # 添加标记点数据项，type_包括'max','min','average'
            data=[
                opts.MarkPointItem(type_="max", name="最大值"),
                opts.MarkPointItem(type_="min", name="最小值"),
            ]
        )
    )
    .add_yaxis(
        series_name='最低气温',
        y_axis=y2,
        markline_opts=opts.MarkLineOpts(                           # 添加标记线数据项
            data=[
                opts.MarkLineItem(type_="average", name="平均值"),            
                opts.MarkLineItem(type_="max", name="最高点")
            ]
        )
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(title="未来一周气温变化"),    # 添加图表标题
        xaxis_opts=opts.AxisOpts(type_="category", boundary_gap=False)   # 将x轴设置为分类轴，数据标签不间隔显示
    )
)

line.render_notebook()
```

通过系列配置项的`AreaStyleOpts`给折线图填充颜色，可以得到面积图：

```python
line = (
    Line()
    .add_xaxis(Faker.choose())     # Faker的choose方法可以随机选择一个分类数据列表
    .add_yaxis(series_name="商家A", y_axis=Faker.values(), is_smooth=True)    # is_smooth=True可以使折线变为平滑曲线
    .add_yaxis(series_name="商家B", y_axis=Faker.values(), is_smooth=True)
    .set_series_opts(
        areastyle_opts=opts.AreaStyleOpts(opacity=0.5),    # 使用默认颜色，透明度为0.5
        label_opts=opts.LabelOpts(is_show=False)           # 不显示数据标签
    )
    .set_global_opts(
        xaxis_opts=opts.AxisOpts(          
            boundary_gap=False    # true:x轴标签两边留白，false：不留白
        )    
    )
)

line.render_notebook()
```

## 层叠多图<a id="4.3"></a>

我们可以将柱状图和折线图合并到一张图上，左Y轴是柱状图的数据刻度，右Y轴是折线图的数据刻度。  
通过Bar对象的`extend_axis`方法可以添加第二个Y轴。

下面是一个层叠多图的示例：

```python
x_data = ["1月", "2月", "3月", "4月", "5月", "6月", "7月", "8月", "9月", "10月", "11月", "12月"]
y1 = [2.0,4.9,7.0,23.2,25.6,76.7,135.6,162.2,32.6,20.0,6.4,3.3]           # 蒸发量
y2 = [2.6,5.9,9.0,26.4,28.7,70.7,175.6,182.2,48.7,18.8,6.0,2.3]           # 降水量
y3 = [2.0, 2.2, 3.3, 4.5, 6.3, 10.2, 20.3, 23.4, 23.0, 16.5, 12.0, 6.2]   # 温度

bar = (
    Bar(init_opts=opts.InitOpts())
    .add_xaxis(x_data)
    .add_yaxis(
        series_name="蒸发量",
        y_axis=y1,
        label_opts=opts.LabelOpts(is_show=False),
    )
    .add_yaxis(
        series_name="降水量",
        y_axis=y2,
        label_opts=opts.LabelOpts(is_show=False),
    )
    .extend_axis(                  # 添加第二个Y轴，用于展示折线图数据
        yaxis=opts.AxisOpts(         
            name="温度",
            type_="value",         # 设置右Y轴的类型为数值轴
            min_=0,                # 设置右Y轴的刻度范围
            max_=25,
            interval=5,            # 每个刻度间隔为5
            axislabel_opts=opts.LabelOpts(formatter="{value} °C"),      # 设置刻度的展示格式为xx°C
        )
    )
    .set_global_opts(
        yaxis_opts=opts.AxisOpts(
            name="水量",
            type_="value",
            min_=0,                # 设置左Y轴的刻度范围
            max_=250,
            interval=50,           # # 每个刻度间隔为50
            axislabel_opts=opts.LabelOpts(formatter="{value} ml"),       # 设置刻度的展示格式
        ),
    )
)

line = (
    Line()
    .add_xaxis(x_data)
    .add_yaxis(
        series_name="平均温度",
        yaxis_index=1,    # 0：使用左Y轴，1：使用右Y轴
        y_axis=y3,        
        label_opts=opts.LabelOpts(is_show=False),
    )
)

bar.overlap(line).render_notebook()      # overlap()方法可以叠加不同的图表
```

## 饼图<a id="4.6"></a>

如果要绘制柱图，可以使用charts模块的Pie对象。

```python
from pyecharts.charts import Pie
```

Pie对象没有`add_xaxis`和`add_yaxis`方法，而是`add`方法，主要参数为：  
* series_name：系列名称
* data_pair：系列数据，格式为 `[(key1, value1), (key2, value2)]`
* color：系列颜色
* radius：饼图的半径，数组的第一项是内半径，第二项是外半径。默认设置成百分比
* center： 饼图的中心（圆心）坐标，数组的第一项是横坐标，第二项是纵坐标。默认设置成百分比
* rosetype：是否展示成南丁格尔图，通过半径区分数据大小，有'radius'和'area'两种模式
* is_clockwise：饼图的扇区是否是顺时针排布

下面是一个饼图的示例：

```python
pie =(
    Pie()
    .add(
        series_name="",
        data_pair=[list(z) for z in zip(Faker.choose(), Faker.values())]
    )
    .set_global_opts(
        legend_opts=opts.LegendOpts(pos_right='right',orient='vertical')     # 设置图例的位置
    )
    .set_series_opts(label_opts=opts.LabelOpts(formatter="{b}: {c}"))     # 设置数据标签的格式
)

pie.render_notebook()
```

通过`radius`参数设置内外半径，可以得到环形图：

```python
pie =(
    Pie()
    .add(
        series_name="",
        data_pair=[list(z) for z in zip(Faker.choose(), Faker.values())],
        radius=["40%","70%"]    # 内半径为40%，外半径为70%
    )
    .set_global_opts(
        legend_opts=opts.LegendOpts(pos_right='right', orient='vertical')     # 设置图例的位置
    )
    .set_series_opts(label_opts=opts.LabelOpts(formatter="{b}: {c}"))     # 设置数据标签的格式
)

pie.render_notebook()
```

通过`rosetype`参数可以得到南丁格尔玫瑰图  
* radius模式：扇区圆心角展现数据的百分比，半径展现数据的大小
* area模式：所有扇区圆心角相同，仅通过半径展现数据大小

```python
v = Faker.choose()
pie = (
    Pie()
    .add(
        series_name="",
        data_pair=[list(z) for z in zip(v, Faker.values())],
        radius=["25%", "60%"],
        center=["25%", "50%"],
        rosetype="radius"
    )
    .add(
        series_name="",
        data_pair=[list(z) for z in zip(v, Faker.values())],
        radius=["25%", "60%"],
        center=["75%", "50%"],
        rosetype="area",
    )
)

pie.render_notebook()
```

## 3D图表<a id="4.8"></a>

pyecharts可以绘制3D柱图、3D折线图、三维地图等3D图表，从而实现三维的可视化效果

以下是一个3D柱图的示例：

```python
import random
from pyecharts.charts import Bar3D

data = [(i, j, random.randint(0, 12)) for i in range(6) for j in range(24)]

bar3D = (
    Bar3D()
    .add(
        series_name="",
        data=[[d[1], d[0], d[2]] for d in data],
        xaxis3d_opts=opts.Axis3DOpts(Faker.clock, type_="category"),
        yaxis3d_opts=opts.Axis3DOpts(Faker.week_en, type_="category"),
        zaxis3d_opts=opts.Axis3DOpts(type_="value"),
    )
    .set_global_opts(
        visualmap_opts=opts.VisualMapOpts(max_=20)
    )
)

bar3D.render_notebook()
```

# 交互效果<a id="5"></a>

交互效果是Echarts的一大特色，上面所有图中的标题、图例都是可以点击的，3D图表可以拖动旋转，除此之外，还有数据区域缩放、tooltip、时间轴等交互操作。

## tooltip和toolbox<a id="5.1"></a>

全局配置项`TooltipOpts`可以添加提示框和指示器，，`ToolboxOpts`可以添加工具箱组件，通过工具箱我们可以在柱状图和折线图之间切换，还可以缩放区域、查看原始数据，保存为图片等等。

```python
x = Faker.week    
y1 = [11, 11, 15, 13, 12, 13, 10]
y2 = [1, -2, 2, 5, 3, 2, 0]

line = (
    Line()
    .add_xaxis(x)
    .add_yaxis(
        series_name='最高气温' ,
        y_axis=y1,
        markpoint_opts=opts.MarkPointOpts(                         
            data=[
                opts.MarkPointItem(type_="max", name="最大值"),
                opts.MarkPointItem(type_="min", name="最小值"),
            ]
        )
    )
    .add_yaxis(
        series_name='最低气温',
        y_axis=y2,
        markline_opts=opts.MarkLineOpts(                           
            data=[
                opts.MarkLineItem(type_="average", name="平均值"),            
                opts.MarkLineItem(type_="max", name="最高点")
            ]
        )
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(title="未来一周气温变化"),    
        tooltip_opts=opts.TooltipOpts(trigger="axis"),    # 配置tooltip，设置为坐标轴触发
        toolbox_opts=opts.ToolboxOpts(is_show=True),      # 配置toolbox，可以看到右上角出现了工具箱
        xaxis_opts=opts.AxisOpts(type_="category", boundary_gap=False)   
    )
)

line.render_notebook()
```

## datazoom<a id="5.2"></a>

当数据量比较大时，可以通过datazoom设置区域缩放

```python
import csv

# 从csv中读取数据
x_data, y_data_rain_fall_amount = [],[]
with open('rainfall.csv','r') as f:
    reader = csv.reader(f)
    next(reader)
    for line in reader:
        x_data.append(line[0])
        y_data_rain_fall_amount.append(line[2])
        
line = (
    Line()
    .add_xaxis(x_data)
    .add_yaxis(
        series_name="降雨量",
        y_axis=y_data_rain_fall_amount,
        label_opts=opts.LabelOpts(is_show=False)
    )
    .set_global_opts(
        tooltip_opts=opts.TooltipOpts(trigger="axis", axis_pointer_type="cross"),
        # 添加数据区域缩放
        datazoom_opts=[
            opts.DataZoomOpts(range_start=0, range_end=100),
            opts.DataZoomOpts(type_="inside", range_start=0, range_end=100),
        ],
        xaxis_opts=opts.AxisOpts(name="时间",type_="category", boundary_gap=False),
        yaxis_opts=opts.AxisOpts(name="降雨量(mm)")
    )
)
line.render_notebook()
```

# 使用主题<a id="6"></a>

pyecharts中内置了许多主题风格，当我们不想自己配置图表的颜色时，可以导入globals模块的ThemeType对象来使用主题。

上面的图形使用的均为默认主题`default`，pyecharts内置的主题有`light`、`dark`、`essos`、`infographic`、`roma`等

下面是将前面的层叠多图套用`MACARONS`主题的效果：

```python
from pyecharts.globals import ThemeType

x_data = ["1月", "2月", "3月", "4月", "5月", "6月", "7月", "8月", "9月", "10月", "11月", "12月"]
y1 = [2.0,4.9,7.0,23.2,25.6,76.7,135.6,162.2,32.6,20.0,6.4,3.3]           # 蒸发量
y2 = [2.6,5.9,9.0,26.4,28.7,70.7,175.6,182.2,48.7,18.8,6.0,2.3]           # 降水量
y3 = [2.0, 2.2, 3.3, 4.5, 6.3, 10.2, 20.3, 23.4, 23.0, 16.5, 12.0, 6.2]   # 温度

bar = (
    Bar(init_opts=opts.InitOpts(theme=ThemeType.MACARONS))
    .add_xaxis(x_data)
    .add_yaxis(
        series_name="蒸发量",
        y_axis=y1,
        label_opts=opts.LabelOpts(is_show=False),
    )
    .add_yaxis(
        series_name="降水量",
        y_axis=y2,
        label_opts=opts.LabelOpts(is_show=False),
    )
    .extend_axis(                  
        yaxis=opts.AxisOpts(         
            name="温度",
            type_="value",         
            min_=0,                
            max_=25,
            interval=5,            
            axislabel_opts=opts.LabelOpts(formatter="{value} °C"),      
        )
    )
    .set_global_opts(
        yaxis_opts=opts.AxisOpts(
            name="水量",
            type_="value",
            min_=0,                
            max_=250,
            interval=50,           
            axislabel_opts=opts.LabelOpts(formatter="{value} ml"),       
        ),
    )
)

line = (
    Line()
    .add_xaxis(x_data)
    .add_yaxis(
        series_name="平均温度",
        yaxis_index=1,    
        y_axis=y3,        
        label_opts=opts.LabelOpts(is_show=False),
    )
)

bar.overlap(line).render_notebook()     
```
